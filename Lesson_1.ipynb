```python
import numpy as np
a = np.array([[1, 6], [ 2, 8], [3, 11], [3, 10], [1, 7]])
a
```

```python
a.shape
```

```python
mean_a = np.mean(a, axis = 0)
```

```python
mean_a
```

```python
a_centered = a - mean_a
```

```python
a_centered
```

```python
a_centered.shape
```

```python
a_centered_sp = np.dot(a_centered[:,0],a_centered[:,1])
a_centered_sp
```

```python
N = a.shape[0]
```

```python
N
```

```python
my_cov=a_centered_sp/(N-1)
```

```python
my_cov
```

```python
m = a.transpose()
m
```

```python
x = np.cov(m)
x
```

```python
if my_cov == x[0, 1]:
  print(True)
```
